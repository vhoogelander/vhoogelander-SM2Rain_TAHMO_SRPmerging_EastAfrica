```python
# Validation years
years = ['2018', '2019', '2022']

data_list = []

# Extract data for each year
for year in years:
    for nstation, ids_data in sm2r_metrics[year].items():
        for ID, metrics_data in ids_data.items():
            for metric, station_values in metrics_data.items():
                for station_id, value in station_values.items():
                    data_list.append([year, nstation, ID, metric, station_id, value])

# Create a DataFrame from the collected data
df = pd.DataFrame(data_list, columns=['year', 'nstations', 'ID', 'metric', 'station', 'value'])
df.loc[df['metric'] == 'bias', 'value'] = -df.loc[df['metric'] == 'bias', 'value']

#Metrics to plot
metrics = ['rmse', 'kge', 'bias', 'spearman']

plt.figure(figsize=(14, 8))

for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='nstations', y='value', hue='year', data=df[df['metric'] == metric])
    if metric == 'kge':
        plt.ylim(-1, 0.5)
        plt.ylabel('[-]')
        plt.title("KLING GUPTA'S EFFICIENCY (KGE)")
        plt.legend([],[], frameon=False)
    if metric == 'rmse':
        plt.ylabel('[mm/day]')
        plt.title('ROOT MEAN SQUARED ERROR (RMSE)')
        plt.legend(title='Year', loc='upper right')
    if metric == 'bias':
        plt.ylabel('[mm/day]')
        plt.title('BIAS')
        plt.legend([],[], frameon=False)
    if metric == 'spearman':
        plt.ylabel('[-]')
        plt.title("SPEARMAN'S CORRELATION")
        plt.legend([],[], frameon=False)
    plt.xlabel('Number of Stations')
    

plt.tight_layout()
```

## Parameter fields

```python
path_to_paramgrids = f'{path_to_sm2rain_runs}/param_grid/'

data_list = []

param_names = ['a', 'b', 'Z*', 'T', 'c']
parameters = ['smoothed_a', 'smoothed_b', 'smoothed_z', 'smoothed_T', 'smoothed_c']

# Loop over the different combinations of letters and numbers
letters = 'ABCDEFGHIJK'
numbers = [10, 25, 50, 75]

for num in numbers:
    for letter in letters:
        # Construct the filename
        filename = f'param_grid_2020_2021_{letter}_{num}.nc'
        file_path = os.path.join(path_to_paramgrids, filename)
        
        if os.path.isfile(file_path):
            ds = xr.open_dataset(file_path)
            
            # Extract data for each parameter
            for param in parameters:
                if param in ds:
                    values = ds[param].values.flatten()
                    for value in values:
                        data_list.append([num, letter, param, value])

# DataFrame from the collected data
df = pd.DataFrame(data_list, columns=['number', 'letter', 'parameter', 'value'])

#Plot results
fig, axs = plt.subplots(2, 3, figsize=(12, 6))

ylabels = {
    'smoothed_a': '[mm/day]',
    'smoothed_b': '[-]',
    'smoothed_z': '[mm]',
    'smoothed_T': '[d]',
    'smoothed_c': '[-]'
}

for i, param in enumerate(parameters):
    ax = axs.flatten()[i]
    sns.boxplot(x='number', y='value', data=df[df['parameter'] == param], ax=ax)
    ax.set_title(param_names[i])
    ax.set_xlabel('Number of stations used in calibration')
    ax.set_ylabel(ylabels[param])

# Hide the last subplot (bottom right)
axs[1,2].set_visible(False)  # Hide the last subplot

plt.tight_layout()
```
